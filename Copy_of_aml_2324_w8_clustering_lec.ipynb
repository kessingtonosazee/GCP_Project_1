```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# https://seaborn.pydata.org/tutorial/aesthetics.html
sns.set(
    style='ticks',
    context='talk',
    font_scale=0.8,
    rc={'figure.figsize': (8,6)}
)
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
penguins = sns.load_dataset('penguins').dropna()
penguins.head()
```

```python
penguins.shape
```

```python
X = penguins.drop(columns=['species', 'island', 'sex'])
```

```python
X.head(1)
```

```python
X = pd.DataFrame(StandardScaler().fit_transform(X), columns=X.columns)
```

```python
X.head(1)
```

```python
kmeans = KMeans(n_clusters=3, n_init='auto').fit(X)
y_pred = kmeans.predict(X)
```

```python
y_pred[:5]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[1].set_title('ground truth')
sns.scatterplot(
    x=X['bill_length_mm'], y=X['bill_depth_mm'], hue=penguins['species'],
    palette='Set2', ax=axs[1]
)
axs[0].set_title('KMeans Clusters on X (333,4) ($K$=3)')
sns.scatterplot(
    x=X['bill_length_mm'], y=X['bill_depth_mm'], hue=y_pred,
    palette='Set1', ax=axs[0]
);
```
