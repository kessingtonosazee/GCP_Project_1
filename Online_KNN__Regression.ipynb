```python
from sklearn.datasets import fetch_california_housing

```

```python
X, y = fetch_california_housing(as_frame=True, return_X_y=True)
```

```python
X.head(1)
```

```python
y.head(1)
```

```python
import numpy as np
import pandas as pd
df = pd.concat([X, y], axis=1)
df.head(1)
```

```python
from seaborn.relational import scatterplot
import seaborn as sns
sns,scatterplot(
    data = df,
    x= 'MedInc',
    y = 'MedHouseVal',
    alpha = 0.2
);
```

```python
dflt5 = df.query('MedHouseVal < 5')

sns,scatterplot(
    data = dflt5,
    x= 'MedInc',
    y = 'MedHouseVal',
    alpha = 0.2
);
```

```python
sns,scatterplot(
    data = dflt5.query("AveRooms<=5"),
    x= 'AveRooms',
    y = 'MedHouseVal',
    alpha = 0.2
);
```

```python
sns.jointplot(
        data=dflt5.query('AveRooms<20'),
        x = 'AveRooms',
        y = 'MedHouseVal',
        kind = 'hex'

    );
```

```python
sns,scatterplot(
    data = dflt5.query("AveRooms<=5"),
    x= 'Latitude',
    y = 'MedHouseVal',
    alpha = 0.2
);
```
