```python
# Import Libraries

from sklearn.datasets import fetch_california_housing
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(
       {"figure.figsize": (6, 4) },
       style='ticks',
       color_codes=True,
       font_scale=0.8
)
%config InlineBackend.figure_format = 'retina'
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn.metrics import PredictionErrorDisplay
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsRegressor

from sklearn.metrics import mean_absolute_error


california.head()
```

```python
# Load and Read dataframe

X, y = fetch_california_housing(return_X_y=True, as_frame=True)
california = pd.concat([X, y], axis=1)
```

```python
# Train-Test Split

X_train, X_test, y_train, y_test = train_test_split(
                X,
                y,
                random_state=0
    )
```

```python
# instantiate and fit model
knnr = KNeighborsRegressor(n_neighbors=11)

knnr.fit(X_train, y_train)
```

```python

# Mean Absolute Error

mean_absolute_error(y_train, knnr.predict(X_train))
```

```python

mean_absolute_error(y_test, knnr.predict(X_test))
```

```python
# Model Analysis (Actual Vs Predicted)

PredictionErrorDisplay.from_estimator(
    knnr,
    X,
    y,
    kind="actual_vs_predicted",
    scatter_kwargs=dict(alpha=0.5)
);
```

```python
# Residual Vs Predicted
PredictionErrorDisplay.from_estimator(
    knnr,
    X,
    y,
    kind="residual_vs_predicted",
    scatter_kwargs=dict(alpha=0.5)
);
```
