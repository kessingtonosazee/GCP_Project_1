```python
# Adjacency Matrix representation in Python


class Graph(object):

    # Initialize the matrix
    def __init__(self, size, directed):
        self.adjMatrix = []
        for i in range(size):
            self.adjMatrix.append([0 for i in range(size)])
        self.size = size
        self.directed = directed
        self.visited_list = [False] * self.size
        self.path = []

    # Add edges
    def add_edge(self, v1, v2):
        if v1 == v2:
            # print("Same vertex %d and %d" % (v1, v2))
            print(f"Same vertex {v1} and {v2}, loops are not allowed!")
            return
        else:
            if self.directed:
                self.adjMatrix[v1][v2] = 1
            else:
                self.adjMatrix[v1][v2] = 1
                self.adjMatrix[v2][v1] = 1

    # Remove edges
    def remove_edge(self, v1, v2):
        if self.adjMatrix[v1][v2] == 0:
            # print("No edge between %d and %d" % (v1, v2))
            print(f"There is no edge between {v1} and {v2}!")
            return
        else:
            if self.directed:
                self.adjMatrix[v1][v2]=0
            else:
                self.adjMatrix[v1][v2] = 0
                self.adjMatrix[v2][v1] = 0

    def __len__(self):
        return self.size

    # Print the matrix
    def print_matrix(self):
        # A = np.matrix(self.adjMatrix)
        # print(A)
        # for row in self.adjMatrix:
        #     for val in row:
        #         print('{:4}'.format(val)),
        #     print
        print('\n'.join([''.join(['{:4}'.format(item) for item in row]) for row in self.adjMatrix]))

    def traverse(self, cur_node):  # recursive

        print(cur_node, end=' ')
        self.visited_list[cur_node] = True
        w = []
        for i in range(self.size):
            if self.adjMatrix[cur_node][i] == 1 and self.visited_list[i] == False:
            # if self.adjMatrix[cur_node][i] == 1:

                w.append(i)
        if len(w) != 0:
            for i in range(len(w)):
                a = w[i]
                if self.visited_list[a] == False:
                    self.traverse(a)
    def reset_states(self):
        self.visited_list = [False] * self.size
    # @pysnooper.snoop()
    def plan(self, cur_node, goal_node):  # recursive

        if cur_node == goal_node:
            self.visited_list[cur_node] = True
            self.path.append(cur_node)
            print(f'visited_list is {self.visited_list}')
            print(f'Path found {self.path}!')
            return self.path

        print(cur_node, end=' ')
        self.visited_list[cur_node] = True
        self.path.append(cur_node)

        print(self.path)
        print(f'visited_list is {self.visited_list}')

        # curr_children = dict()
        w = []
        for i in range(self.size):
            if self.adjMatrix[cur_node][i] == 1 and self.visited_list[i] == False:
            # if self.adjMatrix[cur_node][i] == 1:
                w.append(i)
        # curr_children[cur_node] = w
        print(f'w is {w}')

        if len(w) != 0:
            for i in range(len(w)):
                # print(w)
                a = w.pop(-1)
                result = self.plan(a, goal_node) # before the goal is reached, the plan won't return anything
                # print(result)
                if result is not None:
                    return result



def main():
    num_node = 4
    directed = True
    g = Graph(num_node, directed)
    # six nodes denoted by 0, 1, 2, 3
    # the following codes defines the connectivitiy of the nodes, e.g., g.add_edge(0, 1) means nodes 0 and 1 are connected
    g.add_edge(0, 1)
    g.add_edge(1, 2)
    g.add_edge(0, 3)
    g.add_edge(3, 2)
    g.add_edge(1, 3)
    g.add_edge(3, 1)

    print('==============')

    g.print_matrix()

    print('==============')

    g.traverse(3)
    print('\n')

    g.reset_states()  # traverse will change the states

    print('=========ccc=\n')

    g.plan(0,2)


if __name__ == '__main__':
    main()
```
