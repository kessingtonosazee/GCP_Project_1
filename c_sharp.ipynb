```python
using System;
using System.Collections.Generic;
using System.Linq;
using System.Text;
using System.Threading.Tasks;

namespace ConsoleApp3
{
    class Song
    {
        // members
        private string title, singer;

        // 1st constructor using two argument
        public Song(string title, string name)
        {
            this.title = title;
            this.singer = name;
        }

        // 2nd constructor or alternative constructor using one argument
        public Song(string title)
        {
            this.title = title;
            singer = "unknown";
        }

        // Property
        public string Title
        {
            get { return this.title; }
            set { this.title = value; }
        }


        public string Singer
        {
            get { return this.singer; }
            set { this.singer = value; }
        }

        //Methods
        public void Display()
        {
            Console.WriteLine("The song: {0} by {1}", this.title, this.singer);
        }
    }




}

using System;
using System.Collections.Generic;
using System.Linq;
using System.Text;
using System.Threading.Tasks;

namespace ConsoleApp3
{
    class Album
    {
        // Declare the members
        private string title;
        private Song[] tracks;
        private int numberTracks = 0;


        // Define constructor
        public Album(string title)  //constructor
        {
            this.title = title;
            tracks = new Song[20];  //maximum of 20 songs
        }

        // Define the Properties
        public string Title
        {
            get { return this.title; }
            set { this.title = value; }
        }

        //Define the methods
        public void AddTrack(Song song)
        {
            if (numberTracks < 20)
            {
                tracks[numberTracks] = song;
                numberTracks = numberTracks + 1;
            }
        }

        public void Display()
        {
            for (int i = 0; i < numberTracks; i++)
            {
                tracks[i].Display();
            }
        }

    }
}
```
