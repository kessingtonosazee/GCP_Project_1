### "Is it good practice, to scale transformed categorical data after Target encoding them ?"

* You would have your target in a given scale (say, thousands of GBP). Your predictors might be in different scales (e.g., tens of units). When you are target-encoding, you're making a new predictor the same scale as the target. Why do we scale? To put our predictors in the same scale so as not to make things harder for algorithms such as kNN, regularised Linear Regression (e.g., Lasso, Ridge).
