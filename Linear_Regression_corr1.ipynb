```python
Consumer-Finances-US_Exploring_Data

# Libraries importation
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import wqet_grader

#Read a CSV file to a data frame
df = pd.read_csv("data/SCFP2019.csv.gz")
print("df shape:", df.shape)
df.head()

#Create a subset
mask = df["TURNFEAR"] == 1
df_fear = df[mask]
print("df_fear shape:", df_fear.shape)
df_fear.head()

#Det unique values in a column
age_groups = df_fear["AGECL"].unique()
print("Age Groups:", age_groups)

#Create a series and replace values
agecl_dict = {
    1: "Under 35",
    2: "35-44",
    3: "45-54",
    4: "55-64",
    5: "65-74",
    6: "75 or Older",
}

age_cl = df_fear["AGECL"].replace(agecl_dict)
age_cl.head(10)

#Create a bar chart showing value count
age_cl_value_counts = age_cl.value_counts()
age_cl_value_counts
# Bar plot of `age_cl_value_counts`
age_cl_value_counts.plot(
    kind = "bar",
    xlabel = "Age Group",
    ylabel = "Frequency (count)",
    title = "Credit Fearful : Age Groups"
);

#Create a histogram using pandas
# Plot histogram of "AGE"
df_fear["AGE"].hist(bins = 10)
plt.xlabel("Age")
plt.ylabel("Frequency (count)")
plt.title("Credit Fearful: AAge Distribution");

# Create a series, replace values in a column and crea a hor bar chat
race_dict = {
    1: "White/Non-Hispanic",
    2: "Black/African-American",
    3: "Hispanic",
    5: "Other",
}
race = df_fear["RACE"].replace(race_dict)
race_value_counts = race.value_counts(normalize = True)
# Create bar chart of race_value_counts
race_value_counts.plot(kind = "barh")
plt.xlim((0, 1))
plt.xlabel("Frequency (%)")
plt.ylabel("Race")
plt.title("Credit Fearful: Racial Groups");

#Create a series and convert to a data frame
inccat_dict = {
    1: "0-20",
    2: "21-39.9",
    3: "40-59.9",
    4: "60-79.9",
    5: "80-89.9",
    6: "90-100",
}

df_inccat = (df["INCCAT"]
             .replace(inccat_dict)
             .groupby(df['TURNFEAR'])
             .value_counts(normalize = True)
             .rename("frequency")
             .to_frame()
             .reset_index()
            )

df_inccat

#Use seaburn to create a side by side bar chart
# Create bar chart of `df_inccat`
sns.barplot(
    x = "INCCAT",
    y = "frequency",
    hue = "TURNFEAR",
    data = df_inccat,
    order =inccat_dict.values()
)
plt.xlabel("Income Category")
plt.ylabel("Frequency (%)")
plt.title("Income Distribution: Credit Fearful vs. Non-fearful");

#Cal correlation coefficient of two series
asset_house_corr = df["ASSET"].corr(df["HOUSES"])
print("SCF: Asset Houses Correlation:", asset_house_corr)

#Create a correlation coeff matrix
cols = ["ASSET", "HOUSES", "INCOME", "DEBT", "EDUC"]
corr = df[cols].corr()
#corr.head()
corr.style.background_gradient(axis=None)

# Create Scatter plot
# Create scatter plot of ASSET vs DEBT, df
df.plot.scatter(x = "DEBT", y = "ASSET");
```
